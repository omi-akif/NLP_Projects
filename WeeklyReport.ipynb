````python
WR_KT_df_LC = WR_KT_df[WR_KT_df['robot_code_name'] == 'BD-Daz-LC']
WR_KT_df_EN = WR_KT_df[WR_KT_df['robot_code_name'] == 'BD-Daz-EN']

# WR_KT_df_LC = WR_KT_df
# WR_KT_df_EN = WR_KT_df
````

````python
WR_KT_df_LC = WR_KT_df_LC.reset_index()
WR_KT_df_EN = WR_KT_df_EN.reset_index()
````

### Total session Resolved

#### For EN

````python
WR_KT_df_EN['total_session_resolved'] = 0
for i in range(len(WR_KT_df_EN)):
    WR_KT_df_EN['total_session_resolved'][i] = WR_KT_df_EN['sid_cnt'][i] * WR_KT_df_EN['Session_Resolved_Rate'][i]
````

#### For LC

````python
WR_KT_df_LC['total_session_resolved'] = 0
for i in range(len(WR_KT_df_LC)):
    WR_KT_df_LC['total_session_resolved'][i] = WR_KT_df_LC['sid_cnt'][i] * WR_KT_df_LC['Session_Resolved_Rate'][i]
````

### Total count for switch to live agent

#### For EN

````python
WR_KT_df_EN['total_switch_agent'] = 0
for i in range(len(WR_KT_df_LC)):
    WR_KT_df_EN['total_switch_agent'][i] = WR_KT_df_EN['sid_cnt'][i] * WR_KT_df_EN['Switch_to_Live_Agent_Rate'][i]
````

#### For LC

````python
WR_KT_df_LC['total_switch_agent'] = 0
for i in range(len(WR_KT_df_LC)):
    WR_KT_df_LC['total_switch_agent'][i] = WR_KT_df_LC['sid_cnt'][i] * WR_KT_df_LC['Switch_to_Live_Agent_Rate'][i]
````

### Response Count (Chat)

#### For EN

````python
WR_KT_df_EN['response_count'] = 0
for i in range(len(WR_KT_df_EN)):
    WR_KT_df_EN['response_count'][i] = WR_KT_df_EN['Chats'][i] * WR_KT_df_EN['Survey_Response_Rate'][i]
````

#### For LC

````python
WR_KT_df_LC['response_count'] = 0
for i in range(len(WR_KT_df_LC)):
    WR_KT_df_LC['response_count'][i] = WR_KT_df_LC['Chats'][i] * WR_KT_df_LC['Survey_Response_Rate'][i]
````

### Unsatisfied Count

#### For EN

````python
WR_KT_df_EN['unsatisfied_count'] = 0
for i in range(len(WR_KT_df_EN)):
    WR_KT_df_EN['unsatisfied_count'][i] = WR_KT_df_EN['response_count'][i] * WR_KT_df_EN['Chat_Unsatisfied_Rate'][i]
````

#### For LC

````python
WR_KT_df_LC['unsatisfied_count'] = 0
for i in range(len(WR_KT_df_LC)):
    WR_KT_df_LC['unsatisfied_count'][i] = WR_KT_df_LC['response_count'][i] * WR_KT_df_LC['Chat_Unsatisfied_Rate'][i]
````

### Pivot Table for 

````python
WR_KT_df_EN_pivot =  pd.pivot_table(WR_KT_df_EN, values=['Chats','sid_cnt','total_session_resolved','total_switch_agent','unsatisfied_count'], index= ['Knowledge Name'], aggfunc = np.sum)
WR_KT_df_LC_pivot = pd.pivot_table(WR_KT_df_LC, values=['Chats','sid_cnt','total_session_resolved','total_switch_agent','unsatisfied_count'], index= ['Knowledge Name'], aggfunc = np.sum)
````

````python
WR_KT_df_EN_pivot = WR_KT_df_EN_pivot.sort_values('sid_cnt', ascending=0)
WR_KT_df_LC_pivot = WR_KT_df_LC_pivot.sort_values('sid_cnt', ascending=0)
````

### RR Calculation

````python
WR_KT_df_EN_pivot = WR_KT_df_EN_pivot.reset_index()
WR_KT_df_LC_pivot = WR_KT_df_LC_pivot.reset_index()
````

````python
WR_KT_df_LC_pivot
````

#### For EN

````python
WR_KT_df_EN_pivot['RR'] = 0
for i in range(len(WR_KT_df_EN_pivot)):
    WR_KT_df_EN_pivot['RR'][i] = (WR_KT_df_EN_pivot['total_session_resolved'][i] / WR_KT_df_EN_pivot['sid_cnt'][i]) * 100
````

#### For LC

````python
WR_KT_df_LC_pivot['RR'] = 0
for i in range(len(WR_KT_df_LC_pivot)):
    WR_KT_df_LC_pivot['RR'][i] = (WR_KT_df_LC_pivot['total_session_resolved'][i] / WR_KT_df_LC_pivot['sid_cnt'][i]) * 100
````

### STA Calculation

#### EN

````python
WR_KT_df_EN_pivot['STA'] = 0
for i in range(len(WR_KT_df_EN_pivot)):
    WR_KT_df_EN_pivot['STA'][i] = (WR_KT_df_EN_pivot['total_switch_agent'][i] / WR_KT_df_EN_pivot['sid_cnt'][i])*100
    
````

#### LC

````python
WR_KT_df_LC_pivot['STA'] = 0
for i in range(len(WR_KT_df_LC_pivot)):
    WR_KT_df_LC_pivot['STA'][i] = (WR_KT_df_LC_pivot['total_switch_agent'][i] / WR_KT_df_LC_pivot['sid_cnt'][i])*100
    
````

### UR Calculation

#### EN

````python
WR_KT_df_EN_pivot['UR'] = 0
for i in range(len(WR_KT_df_EN_pivot)):
    WR_KT_df_EN_pivot['UR'][i] = (WR_KT_df_EN_pivot['unsatisfied_count'][i] / WR_KT_df_EN_pivot['Chats'][i])*100
````

#### LC

````python
WR_KT_df_LC_pivot['UR'] = 0
for i in range(len(WR_KT_df_LC_pivot)):
    WR_KT_df_LC_pivot['UR'][i] = (WR_KT_df_LC_pivot['unsatisfied_count'][i] / WR_KT_df_LC_pivot['Chats'][i])*100
````
