```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x=np.linspace(1,5,10)
```

```python
y=x**2
```

```python
plt.plot(x,y, "red",linewidth=1,markersize=10, marker="+")
plt.xlabel("XLabel")
plt.ylabel("Y Label")
plt.title("First Chart")
plt.show()
```
