### Pivot Table for 

````python
A_KT_df[['Date','Total Sessions','total_session_resolved','total_switch_agent','unsatisfied_count']].head(2)
````

````python
A_KT_df =  pd.pivot_table(A_KT_df, values=['Total Sessions', 'total_session_resolved','total_switch_agent','unsatisfied_count'], index= ['Week_Number'], aggfunc = np.sum)
# WR_KT_df_LC_pivot = pd.pivot_table(WR_KT_df_LC, values=['Chats','sid_cnt','total_session_resolved','total_switch_agent','unsatisfied_count'], index= ['Knowledge Name'], aggfunc = np.sum)
````

````python
A_KT_df
````

````python
A_KT_df['RR%'] = (A_KT_df['total_session_resolved'] / A_KT_df['Total Sessions']) * 100
A_KT_df['RR%'] = A_KT_df['RR%'].round(decimals = 2)
````

````python
A_KT_df['SLAR%'] = (A_KT_df['total_switch_agent'] / A_KT_df['Total Sessions']) * 100
A_KT_df['SLAR%'] = A_KT_df['SLAR%'].round(decimals = 2)
````

````python
A_KT_df['UR%'] = (A_KT_df['unsatisfied_count'] / A_KT_df['Total Sessions']) * 100
A_KT_df['UR%'] = A_KT_df['UR%'].round(decimals = 2)
````
